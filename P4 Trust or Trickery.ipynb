```python
import pandas as pd

url = "https://raw.githubusercontent.com/fivethirtyeight/guns-data/master/full_data.csv"
df = pd.read_csv(url)
```

```python
df.head(20)
```

```python
import plotly.express as px



intent_counts = df['intent'].value_counts()[['Suicide','Homicide','Accidental','Undetermined']].reset_index()
intent_counts.columns = ['Intent', 'Count']


fig = px.bar(
    intent_counts,
    x='Intent',
    y='Count',
    title='Counts of Gun Deaths by Intent',
    color='Intent',
    text='Count'
)

fig.update_layout(
    xaxis_title="Intent Category",
    yaxis_title="Number of Deaths",
    showlegend=False
)

fig.update_traces(textposition='outside')

fig.show()
```

```python
import plotly.express as px

# 1 Omit suicides 
df_no_suicide = df[df['intent'] != 'Suicide']

# 2 Use percentages of the remaining categories 
mis_counts = (
    df_no_suicide['intent']
    .value_counts(normalize=True)   
    .reset_index()
)
mis_counts.columns = ['Intent', 'Share']

fig_trick = px.bar(
    mis_counts,
    x='Intent',
    y='Share',
    title='What really causes gun deaths? ',
    text=(mis_counts['Share'] * 100).round(1).astype(str) + '%'
)

# 3 Truncate the y-axis 
fig_trick.update_yaxes(
    title_text='Share of gun deaths ',
    range=[0.6, 1.0]   
)

fig_trick.update_layout(
    showlegend=False,
    xaxis_title="Intent",
)

fig_trick.update_traces(textposition='outside')
fig_trick.show()
```
